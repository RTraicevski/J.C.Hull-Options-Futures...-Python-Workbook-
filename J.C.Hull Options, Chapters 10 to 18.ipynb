## Chapter 10 Mechanics of Options Markets

A call option gives the holder of the option the right to buy an asset by a certain date for a certain price. 

A put option gives the holder the right to sell an asset by a certain date for a certain price. 

The date specified in the contract is known as the expiration date or the maturity date.The price specified in the contract is known as the exercise price or the strike price.

## Payoffs

### Call Option
#### Long Position
$$max(S_{T} - K,0) $$
#### Short Position
$$min(K - S_{T},0) $$

### Put Option
#### Long Position
$$max(K - S_{T},0) $$
#### Short Position
$$min(S_{T} - K,0) $$


```python
## Payoffs

K = 10 # Strike Price

spot_price = 5
ST = np.arange(0.1*spot_price,5*spot_price,1)


### Call Option Long or Short

def callPayoff(ST,K,):
    Long = np.where(ST > K, ST - K,0)
    Short = np.where(ST > K, K - ST, 0)
    return Long, Short

CLongPayoff = callPayoff(ST,K)[0]
CShortPayoff = callPayoff(ST,K)[1]

fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.plot(ST,CLongPayoff,label='Long Position: Call Option')
plt.xlabel("Asset prices")
plt.ylabel("Payoff")
plt.legend(loc='upper right')

           
fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.plot(ST,CShortPayoff,label='Short Position: Call Option')
plt.xlabel("Asset prices")
plt.ylabel("Payoff")
plt.legend(loc='upper right')
plt.show()
    
    
### Put Option Long or Short

def PutPayoff(ST,K,):
    Long = np.where(K > ST, K - ST,0)
    Short = np.where(K > ST, ST - K, 0)
    return Long, Short

PLongPayoff = PutPayoff(ST,K)[0]
PShortPayoff = PutPayoff(ST,K)[1]

fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.plot(ST,PLongPayoff,label='Long Position: Put Option')
plt.xlabel("Asset prices")
plt.ylabel("Payoff")
plt.legend(loc='upper right')

           
fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.plot(ST,PShortPayoff,label='Short Position: Put Option')
plt.xlabel("Asset prices")
plt.ylabel("Payoff")
plt.legend(loc='upper right')
plt.show()
    
```

## Chapter 11 Properties of Stock Options

There are six factors affecting the price of a stock option:
1. The current stock price, S0
2. The strike price, K
3. The time to expiration, T
4. The volatility of the stock price,
5. The risk-free interest rate, r
6. The dividends that are expected to be paid

#### Upper and Lower Bounds for Option Prices

The following Bounds must hold, otherwise there will be arbitrage possibilities

$S_{0}$:Current Stock Price \
$S_{T}$: Stock price on expiration date \
$K$: Strike Price of option\
$r$: Continuously Compounded Risk Free Rate, r > 0 

$c$: Value of Call Option to buy one share\
$p$: Value of Put Option to sell one share

#### Upper Bound
Stock price is an upper bound to either a Euro or Am Call option,
$$c \leq S_{0}$$ \
For an Am Put, the option price can never be worth more than the strike price,
$$p \leq K$$
The Euro Put can never be worth more than the strike price at maturity, then discounting for to have it in terms of today,
$$ p \leq Ke^{-rT}$$

#### Lower Bound
Price of a Euro Call option on a non dividend paying stock is,
$$c \geq S_{0} - Ke^{-rT}$$
$$c \geq max(S_{0} - Ke^{-rT}, 0) $$

Price of a Euro Put option on a non dividend paying stock is,
$$p \geq Ke^{-rT} - S_{0} $$
$$p \geq max(Ke^{-rT} - S_{0}, 0) $$

#### Put Call Parity (Euro)
Assuming the Put and Call Option have the same strike price and maturity, and that any stock pays no dividends,
$$c + Ke^{-rT} = p + S_{0}$$

#### American Put Call Parity
When there are no dividends, (Put call parity only holds for Euro Options)
$$S_{0} - K \leq c - p \leq S_{0} - Ke^{-rT}$$



#### Calls on a non Dividend paying stock
It is never optimal to exercise an American call option on a non dividend paying stock before the expiration date

Because the owner of an American Call has the right to early exercise,
$$ c_{Am} \geq c_{euro}$$
Hence,
$$c_{Am} \geq c_{euro} \geq S_{0} - Ke^{-rT}$$

This means when there are no dividends, the price of an American call is equal to a European Call option

#### Puts on a non Diviend paying stock
It can be optimal to early exercise an American Put option on a non dividend paying stock, as it may be optimal for an investor to exercise early to realise the strike price immediately. Therefore, the bounds are,
$$p_{Am} \geq max(K - S_{0}, 0)$$
$$max(K - S_{0} \leq p_{Am} \leq K$$




#### Effect of Dividends (Euro)
$D$: Present Value of the Dividend

#### Lower Bound
Call Option
$$c \geq max(S_{0} - D - Ke^{-rT}, 0) $$

Put Option
$$p \geq max(D + Ke^{-rT} - S_{0}, 0) $$

#### Put Call Parity
$$c_{Euro} + D + Ke^{-rT} = p_{Euro} + S_{0}$$
$$S_{0} - D - K \leq c_{Am} - p_{Am} \leq S_{0} - Ke^{-rT}$$

## Chapter 14 Wiener Processes and Ito's Lemma

#### Markov Property
A Markov process is a particular type of stochastic process where only the current value of a variable is relevant for predicting the future

The Markov property of stock prices is consistent with the weak form of market
efficiency. This states that the present price of a stock impounds all the information contained in a record of past prices

#### Wiener Process (Brownian Motion)

Property 1. The change $\Delta z$ during a small period of time $\Delta t$, 
$$\Delta z = \epsilon \sqrt{\Delta t}$$
Where $\epsilon$ has a standard normal distribution

Property 2. $\Delta z$ for any two different short intervals of time,$\Delta t$ 
are independent.
$$E[\Delta z]=0$$
$$Var[\Delta z]=\Delta t$$

$$E[z(T) - z(0)] = 0$$
$$V[z(T) - z(0)] = T$$

#### Generalised Wiener Process
Where z is the Wiener process described above,
$$ dx = adt + bdz$$

Without the dz term, dx = adt implies, $x = x_{0} + at$
Thus we have that,
$$\Delta x = a\Delta t + b\epsilon \sqrt{\Delta t}$$

With the following properties,
$$E[\Delta x]=a\Delta t$$
$$Var[\Delta x]=b^{2}\Delta t$$

#### Process for a stock price
Using that the drift rate is the epected return of the stock and is constant, and assuming uncertainty in a short period of time is constant, independent of stock price,
$$dS = \mu Sdt + \sigma Sdz$$
Or in Discrete Time Model Format,
$$\Delta S = \mu S\Delta T + \sigma S\sqrt{\Delta t}$$

#### Ito's Lemma
Supposing the value of a variable x follows the Ito Process,
$$dx = a(x,t)dt + b(x,t)dz$$
where dz is a Wiener process
Then Ito's Lemma holds that a function G of x and t follows the process,
$$ dG = (\frac{\partial G}{\partial x} a + \frac{\partial G}{\partial t} +  \frac{1}{2} \frac{\partial^2  G}{\partial x^2} b^2)dt + \frac{\partial G}{\partial x}bdz \qquad (1)$$

#### Forward Contract using Ito's Lemma
$$F_{0} = S_{0}e^{rT}$$
$$F = Se^{r(T-t)}$$\
$$\frac{\partial F}{\partial S} = e^{r(T-t)},\quad \frac{\partial^2  F}{\partial s^2}=0,\quad \frac{\partial F}{\partial t}=-rSe^{r(T-t)}$$

Then using (1), the process for F is given by
$$dF = [e^{r(T-t)}\mu S -rSe^{r(T-t)}]dt + e^{r(T-t)}\sigma Sdz$$

## Chapter 15 The Black Scholes Merton Model

### Lognormal Property of Stock Prices
(Remember that stock prices are modelled as lognormal, while returns are modelled as normal)

From the following
$$\Delta S = \mu S\Delta T + \sigma S\sqrt{\Delta t}$$
We can conclude that,
$$\frac{\Delta S}{S} = \mu \Delta T + \sigma \sqrt{\Delta t} \sim N(\mu \Delta t,\sigma^2 \Delta t)$$
This means,
$$lnS_{T} \sim N[lnS_{0}+(\mu -\frac{\sigma^2 }{2})T, \sigma^2 T]$$
$$E(S_{T})=S_{0}e^{\mu T}$$
$$Var(S_{T})=S_{0}^{2}e^{2\mu T}(e^{\sigma^2 T}-1)$$
##### (This can be used to find confidence intervals for a possible stock price at some time in the future)

### The Distribution of the Rate of Return
Taking the continuously compounded rate of return,
$$S_{T}=S_{0}e^{xT}$$
$$x=\frac{1}{T}ln\frac{S_{T}}{S_{0}}$$
Then using
$$ln\frac{S_{T}}{S_{0}} \sim N[(\mu -\frac{\sigma^2 }{2})T, \sigma^2 T]$$
We have that the rate of return is normally distributed,
$$ x \sim N[\mu -\frac{\sigma^2 }{2},\frac{\sigma^2 }{T}]$$
##### Dont forget, $E(x) = \mu -\frac{\sigma^2 }{2}$ and $Var(x)=\frac{\sigma^2 }{T}$

### Black Scholes Merton PDE

1. The stock price follows the process developed in Chapter 14 with $\mu$ and $\sigma$ constant.

4. There are no dividends during the life of the derivative.
5. There are no riskless arbitrage opportunities.
6. Security trading is continuous.
7. The risk-free rate of interest, r, is constant and the same for all maturities.
